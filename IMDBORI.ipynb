```python
from flask import Flask, render_template
import pandas as pd

# Load your dataset into a Pandas DataFrame (replace 'path_to_your_cleaned_data.csv' with your data path)
data = pd.read_csv('IMDB-Movie-Data.csv')

# Create an instance of the Flask app
app = Flask(__name__)

# Define a route to render the data
@app.route('/')
def index():
    return render_template('index.html', data=data.to_html(classes='data', header="true", index=False))

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True, port=5001)  # Change port number if needed

```
